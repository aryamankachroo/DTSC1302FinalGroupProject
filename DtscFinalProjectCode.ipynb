```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.hist(figsize=(12,8))
plt.show()
```

```python
df.columns = ['age','workclass','fnlwgt','education','educationnum','maritalstatus','occupation','relationship','race','sex','capitalgain', 'capitalloss', 'hoursperweek','nativecountry', 'salary']
df.head()
```

```python
df.head()
```

```python

plt.figure(figsize = (8,6))
numeric = df.select_dtypes(include=['number'])
matrix = numeric.corr()
sns.heatmap((matrix), annot = True, cmap = 'coolwarm', linewidths = 0.5)
plt.title('Heatmap of Correlation')
plt.show()
```

```python
df['salary'] = df['salary'].replace({' <=50K': 0, ' >50K': 1})
#df['salary'] = df['salary'].astype(int)
sns.displot(x = df['salary'])
```

```python
categorical = df.select_dtypes(include=['object'])
sns.barplot(x = 'workclass', y = 'salary', data = df, palette = 'tab10')
plt.title('Workclass vs Salary')
plt.xticks(rotation = 90)
```
